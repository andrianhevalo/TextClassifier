from Okean Elzy or Tartak .

```python
import nltk
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from nltk.tokenize import TweetTokenizer
from stop_words import get_stop_words
from sklearn.feature_extraction.text import CountVectorizer
import json
import string
```

### Read and preprocess the datasets.

1. Creating a list of stop words and punctuation that don't impact on model performance
2. Creating a data frame
3. Data Encoding

#### Creating a list of stop words and punctuation that don't impact on model performance

```python
stop_words = get_stop_words('ukrainian') + [string.ascii_lowercase]
punctuation = [',', ':', ';', '!', '?', '–', 'й', 'а', "'", '"', 'приспів', '(', ')', '.', '..', '...']
stop_words += punctuation
stop_words
```

#### Creating a data frame and feature engeneering

In any regression problem it is essencial to select features that would boost the model performance.

```python
def create_dataframe(path, labels):
    """
    Creates a data frame with corrensponding columns:
    lyric - song lyrics,
    title = song title, 
    word_number - word count in song lyrics, 
    Кількість Букви Р,
    author
    """
    lyrics = []
    authors = []
    titles = []
    words_number = []
    r_numbers_in_songs = []
    for key, value in labels.items():
        temp = key.split(".")
        #  reading the file
        with open(path + "\\texts\\" + temp[0] + ".txt", "r", encoding = 'utf-8') as file:
            lyric = file.read().replace("’","").strip("\n").split(" ")
            lyrics.append(" ".join(lyric))
            words_number.append(len(lyric))
            r_numbers_in_songs.append(" ".join(lyric).count('р'))
        authors.append(value['artist'])
        titles.append(value['title'])
            
    df = pd.DataFrame({"lyric": lyrics, "title": titles, "word_number": words_number, "Кількість Букви Р": r_numbers_in_songs, "author": authors})
    return df
```

```python
#  train df

jsonfile = open("train\\labels.json", "r", encoding='utf8').read()
labels = json.loads(jsonfile)
df = create_dataframe("train", labels)
df.head()
```

```python
#  test df

jsonfile = open("test\\labels.json", "r", encoding='utf8').read()
labels = json.loads(jsonfile)
df_test = create_dataframe("test", labels)
df_test.head()
```

### Conclusions

I run three models for this classification problem and achieved highest accuracy (~93%). The problem was to select relevant featutes for songs author detection. I tried some like number of stop words, total number of words, etc. Most relevant are - Кількість Букви Р (boosted model performance nearly by 20%) and total word count in song lyrics. Short explanation on my choice:
#### 1. Okean Elzy's songs have tendency to have less "р" letter count due to main vocalist capabilities of pronunciation.
#### 2. Tartak's songs have some rap vibes, so their songs tend to have more words.
