#### Importing all packages 

```python
import pandas as pd
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
import sklearn as sk
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, KFold, cross_val_score, GridSearchCV
from sklearn.metrics import classification_report, f1_score, accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn import metrics
import pandas as pd
import os
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

### Loading data and droping columns for models



```python
train_data = pd.read_csv("/Users/andrianhevalo/Desktop/HW_1/homework1cancer-classification/train.csv")
test_data = pd.read_csv("/Users/andrianhevalo/Desktop/HW_1/homework1cancer-classification/test.csv")
train_data
```

### Data preprocessing: cleaning missing values and fixing non-numeric values

```python
models_accuracy = list()

# x and y for models
x  = train_data.drop(['id', 'diagnosis'], axis=1)
y = train_data['diagnosis']


train_data.isnull().sum()
train_data.isna().sum()
```

### Splitting the dataset into three subsets: train, CV, test using 60-20-20% rule 

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size = 0.25, random_state=42)
```

### Data normalization
Normalization is required only when features have different ranges.

```python
scaler = StandardScaler()
x_scaled = pd.DataFrame(scaler.fit_transform(x))
x_scaled.columns = x.columns

test_data_scaled = pd.DataFrame(scaler.transform(test_data.drop(['id'], axis=1)))
test_data_scaled.columns = test_data.drop(['id'], axis=1).columns
test_data_scaled.head()
test_data_scaled
```

### Training Logistic Regression with scaled data

```python
kfold = KFold(n_splits=5, random_state=40)

model_log = LogisticRegression()

model_log.fit(x_train, y_train)

y_pred = model_log.predict(x_test)

models_accuracy.append(("Logistic Regression",  cross_val_score(model_log, x_scaled, y, cv=kfold).mean() * 100))

# Plot confusion matrix
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
p = sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')

print(classification_report(y_test,y_pred))
```

### Correlations with dependent variable y

```python
train_data.drop("diagnosis", axis=1).apply(lambda x: x.corr(train_data.diagnosis))
```

### correlation matrix heatmap for the data set
The presence of a highly-correlated feature won't affect the final prediction by the model. However, the estimates of the coefficients may not make sense.
A regression coefficient measures the change in Y per unit change in that predictor, will all predictors held fixed. But correlated predictors change together.

```python
corr = train_data.corr()

ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```

### Hyperparameter Grid
#### Based on our the correlation of our data it randomly computes parameters for Logistic Regression to increase its performance
Grid search is an approach to hyperparameter tuning that will methodically build and evaluate a model for each combination of algorithm parameters specified in a grid.

```python
penalty = ['l1', 'l2']
C = [0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000]
class_weight = [{1:0.5, 0:0.5}, {1:0.4, 0:0.6}, {1:0.6, 0:0.4}, {1:0.7, 0:0.3}]
solver = ['liblinear', 'saga']

param_grid = dict(penalty=penalty,
                  C=C,
                  class_weight=class_weight,
                  solver=solver)

grid = GridSearchCV(estimator=model_log,
                    param_grid=param_grid,
                    scoring='roc_auc',
                    verbose=1,
                    n_jobs=-1)

grid_result = grid.fit(x_scaled, y)

grid_pred = grid.predict(x_scaled)

print('Best Score: ', grid_result.best_score_)
print('Best Params: ', grid_result.best_params_)
print(classification_report(grid_pred, y))
models_accuracy.append(("Hyperparameters Logistic Regression", grid_result.best_score_ * 100))
```

### Training Support Vector Machine

```python
model_svc = SVC(kernel='linear')

model_svc.fit(x_train, y_train)

y_pred = model_svc.predict(x_test)

models_accuracy.append(("SVM", accuracy_score(y_pred, y_test) * 100))
      
# Plot confusion matrix
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
p = sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
      
print(classification_report(y_test,y_pred))
```

### Training decision trees

```python
from sklearn import tree

classification_tree = tree.DecisionTreeClassifier()

classification_tree = classification_tree.fit(x_train, y_train)

#Predict the response for test dataset
y_pred = classification_tree.predict(x_test)

models_accuracy.append(("Decision Tree", accuracy_score(y_pred, y_test) * 100))

# Plot confusion matrix
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
p = sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')

print(classification_report(y_test,y_pred))
```

## Conclusion

###### So finally we have built our classification models and we can see that Hyperparameters Logistic Regression algorithm gives almost highest possible accuracy (99.3%). 
To choose a model it is always needed to analyze the dataset and then apply a machine learning model. 
Colinearity and outliers tampers the accuracy of Logistic Regression model. In our case data has a little noise which impact on the final model score.
Linear SVM handles outliers better, as it derives maximum margin solution.
For Decision Trees preprocessing isn't needed on data.
Since our data has high dimensional space SVM is effective in this case. There are a lot of reasons why we should use cross validation for prediction like using all data, parameters fine-tuning but in our case cross validation gives low results due to noisy data ans outliers. 
In this case it is not sufficient to use all data, better to prepare it and clean and then apply Hyperparameters Logistic Regression to reach high accuracy!
